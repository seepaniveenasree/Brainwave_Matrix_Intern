Task1:-Create a sale data analysis of any commercial store.The dataset is from github(Supermarket store branches sales analysis)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path="C:\\Users\\seepa\\Datasets\\Stores.csv"
```

```python
# Load the dataset
df= pd.read_csv(file_path)
```

```python
#Explore the dataset
# Display the first 5 rows
df.head()
```

```python
# Get basic information about the dataset
df.info()
```

```python
# Get summary statistics
df.describe()
```

```python
df.shape
```

```python
#Clean the dataset
# Check for missing values
df.isnull().sum()
```

```python
#No null values
# Correlation analysis
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Visualizations
plt.figure(figsize=(10, 6))
sns.histplot(df['Store_Sales'], bins=30, kde=True)
plt.title('Distribution of Store Sales')
plt.xlabel('Store Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Store Area vs Sales scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Store_Area'], y=df['Store_Sales'])
plt.title("Store Area vs Store Sales")
plt.show()
```
